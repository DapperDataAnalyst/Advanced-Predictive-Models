# Peer-Graded SVD and PCA

### Part (a)

```python
import numpy as np
import matplotlib.pyplot as plt

# Get data
data = np.genfromtxt('PCAData.csv', delimiter=',')

# Remove rows with np.nan
mask = np.any(np.isnan(data), axis=1)
data = data[~mask]

# Extract columns of interest
x = data[:,1]
y = data[:,2]

# Create a scatter plot
plt.scatter(x, y)
plt.title('Scatter Plot')

# Show the plot
plt.show()
```

The points form a cloud centered approximately around an x value of 3 and a y value of -1.

The cloud of points seems to exhibit most of its variation approximately in the direction of the line y = x.

### Part (b)

```python
from sklearn.decomposition import PCA

# Create a PCA instance and fit the data
pca = PCA(n_components=2)
fit = pca.fit_transform(data[:,1:3])

# Access eigenvalues and eigenvectors
eigenvalues = pca.explained_variance_
eigenvectors = pca.components_

# Print eigenvalues and eigenvectors
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:")
print(eigenvectors)
```

From this analysis, we find eigenvalue $\lambda_1 = 8.17$ with corresponding eigenvector [-0.81, -0.58] and eigenvalue $\lambda_2 = 1.73$ with corresponding eigenvector [-0.58, 0.81].

### Part (c)

```python
import math

mean_x = np.mean(data[:,1])
mean_y = np.mean(data[:,2])

# Create a scatter plot
plt.scatter(x, y)
plt.arrow(x=mean_x, y=mean_y, dx=eigenvectors[0,0]*math.sqrt(eigenvalues[0]),dy=eigenvectors[1,0]*math.sqrt(eigenvalues[0]), width = 0.15, color='red')
plt.arrow(x=mean_x, y=mean_y, dx=eigenvectors[0,1]*math.sqrt(eigenvalues[1]), dy=eigenvectors[1,1]*math.sqrt(eigenvalues[1]), width = 0.15, color='red')
plt.title('Scatter Plot')

# Show the plot
plt.show()
```

The eigenvectors are orthogonal to one another.

The longer vector points diagonally, approximately along the path of the line y = x. This demonstrates that this longer eigenvector is the PC1 direcetion, with the shorter vector being the PC2 direction.
