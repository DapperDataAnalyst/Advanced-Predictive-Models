### Q1

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Replace 'your_csv_file.csv' with the path to your CSV file
df = pd.read_csv('AdjMatrix.csv', header=None)

# Create the heatmap using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(df, cmap='coolwarm', annot=True, fmt='d', cbar=False)
plt.title('Adjacency Matrix Heatmap')
plt.xlabel('Nodes')
plt.ylabel('Nodes')
plt.show()
```

### Q2

```python
# Check if the DataFrame is symmetrical
is_symmetrical = df.equals(df.T)

if is_symmetrical:
    print("The DataFrame is undirected.")
else:
    print("The DataFrame is directed.")
```
